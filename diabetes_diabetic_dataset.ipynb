## Importing essential libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

## Data Preparation

### Loading the dataset

```python
df = pd.read_csv('dataset_diabetes_diabetic_data.csv')
df.head()
```

### Checking number of features and data present in the dataset 

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include="all")
```

### Checking null values in dataset

```python
df.isnull().sum()
```

### Checking unique values per column in dataset

```python
for i in df.columns:
    print(i,df[i].nunique())
```

### Dropping the unnecessary data columns and rows

```python
df.drop(columns=['patient_nbr','diag_1','diag_2','diag_3'],axis=1,inplace=True)
df
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
count = 0
for index, row in df.iterrows():
    if row['race'] == '?':
        count += 1
print(count)
```

```python
df.drop(df.loc[df['race']=='?'].index, inplace=True)
df
```

## Splitting data into Categorical & Numerical

```python
df_cat = pd.DataFrame()
df_num = pd.DataFrame()
for y in df.columns:
    if df[y].dtype == object:
        df_cat[y] = df[y]
    else:
        df_num[y] = df[y]
```

### Finding Correlation on numerical data 

```python
corr = df_num.corr()
corr
```

### Heatmap of correlation

```python
plt.figure(figsize = (20,10))
sns.heatmap(corr,annot = True,cmap='rainbow')
plt.show()
```

## Label Encoding

```python
for i in df_cat.columns:
    print(i,"\n",df[i].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
for i in df_cat.columns:
    df[i] = lb.fit_transform(df[i])
df
```

## Splitting of dataset

### Separating target and other features

```python
X = df.iloc[:,:-1]
X
```

```python
Y = df.iloc[:,15:]
Y
```

### Testing - Training Data Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.4,random_state=0)
```

## Applying DecisionTreeClassifier

### Using Gini Index

```python
ginindex = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_leaf=3,random_state=100)
ginindex.fit(X_train,Y_train)
y_pred = ginindex.predict(X_test)
accuracy = accuracy_score(Y_test,y_pred)
cm = confusion_matrix(Y_test,y_pred,labels=ginindex.classes_)
print(cm,'\n',accuracy,'\n',classification_report(Y_test,y_pred))
```

```python
tree.plot_tree(ginindex)
```

### Using Entropy

```python
entropy = DecisionTreeClassifier(criterion='entropy',max_depth=5,min_samples_leaf=3,random_state=100)
entropy.fit(X_train,Y_train)
y_pred_entropy = entropy.predict(X_test)
cm = confusion_matrix(Y_test,y_pred_entropy,labels=entropy.classes_)
accuracy = accuracy_score(Y_test,y_pred)
print(cm,'\n',accuracy)
print(classification_report(Y_test,y_pred))
```

```python
tree.plot_tree(entropy)
```

### Using GridSearchCV 

```python
tree_para = {'criterion':['gini','entropy'],'max_depth':np.arange(3, 15)}
clf = GridSearchCV(DecisionTreeClassifier(), tree_para, cv=5)
clf.fit(X_train, Y_train)
```

```python
print(clf.best_params_)
print(clf.score(X_test,Y_test))
```

### AUC-ROC Curve

```python
Y_score = clf.predict_proba(X_test)[:,1]
Y_score
```

```python
false_positive_rate, true_positive_rate, threshold = roc_curve(Y_test, Y_score)
print('roc_auc_score for DecisionTree: ', roc_auc_score(Y_test, Y_score))
```

```python
plt.subplots(1, figsize=(10,10))
plt.title('Receiver Operating Characteristic - DecisionTree')
plt.plot(false_positive_rate, true_positive_rate)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

## Applying RandomForest

```python
rf = RandomForestClassifier(n_estimators=100,random_state=0)
rf.fit(X_train,Y_train)
y_pred = rf.predict(X_test)
cm = confusion_matrix(Y_test,y_pred,labels=rf.classes_)
print(accuracy_score(Y_test,y_pred),"\n",cm)
print(classification_report(Y_test,y_pred))
```

### AUC-ROC Curve

```python
Y_score = rf.predict_proba(X_test)[:,1]
Y_score
```

```python
false_positive_rate, true_positive_rate, threshold = roc_curve(Y_test, Y_score)
print('roc_auc_score for RandomForest: ', roc_auc_score(Y_test, Y_score))
```
